# Beauty Parlor System

***In a real-world UI, users would select services using buttons or dropdown menus, but in this program, inputs are typed manually.***

```python
# Services dictionary: service name mapped to price
services = {
    "haircut": 20,
    "facial": 40,
    "massage": 60,
    "manicure": 25,
    "pedicure": 30
}

# Stores all bookings
bookings = []

# Displays available services
def show_services():
    print("\nAvailable Services:")
    for name, price in services.items():
        print(f"{name} - ${price}")

# Handles booking with multiple services + receipt + validation
def book_appointment():
    try:
        name = input("\nName: ").strip().title()
        show_services()

        # Multiple services input
        service_input = input("Choose services (comma separated): ").lower()
        selected_services = [s.strip() for s in service_input.split(",")]

        time = input("Enter time (e.g., 3 pm): ").strip().lower()

        # Validate time format
        if "am" not in time and "pm" not in time:
            print("Invalid time format.")
            return

        # Prevent double booking
        for b in bookings:
            if b["time"] == time:
                print("This time slot is already booked.")
                return

        # Validate services and calculate total
        total = 0
        valid_services = []

        for s in selected_services:
            if s in services:
                valid_services.append(s)
                total += services[s]
            else:
                print(f"Invalid service: {s}")
                return

        # Save booking
        booking = {
            "name": name,
            "services": valid_services,
            "time": time,
            "total": total
        }

        bookings.append(booking)

        # Print receipt
        print("\n--- RECEIPT ---")
        print(f"Name: {name}")
        print(f"Time: {time}")
        print("Services:")
        for s in valid_services:
            print(f"- {s.title()} (${services[s]})")
        print(f"Total Cost: ${total}")
        print("----------------")
        print("Booking confirmed!")

    except:
        print("Something went wrong. Try again.")

# Displays all bookings
def show_bookings():
    try:
        if not bookings:
            print("\nNo bookings available.")
            return

        print("\nAll Bookings:")
        for b in bookings:
            print(f"{b['name']} | {b['services']} | {b['time']} | ${b['total']}")

    except:
        print("Error displaying bookings.")

# Main program loop
while True:
    print("\n1. Book  2. View  3. Exit")

    try:
        choice = input("Option: ").strip()

        if choice == "1":
            book_appointment()
        elif choice == "2":
            show_bookings()
        elif choice == "3":
            print("Goodbye!")
            break
        else:
            print("Invalid choice")

    except:
        print("Invalid input.")
```
